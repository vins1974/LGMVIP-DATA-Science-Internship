##Image to Pencil Sketch with Python

###Importing all libraries

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

###loading image

```python
image=cv2.imread("/content/puppy-development.jpg",cv2.IMREAD_UNCHANGED)
```

```python
cv2_imshow(image)
```

###CONVERTING THE IMAGE INTO GRAY SCALE IMAGE

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

###INVERTING THE IMAGE

```python
image_invert = cv2.bitwise_not(image_gray)
cv2_imshow(image_invert)
#invert_image=255-grey_image
```

Converting inverting image to blur image

```python
image_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

## FINAL PENCIL SKETCH

```python
def dogV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dogV2(image_gray, image_smoothing)
cv2_imshow(final_img)
```
